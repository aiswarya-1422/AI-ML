```python
data = pd.read_csv('/content/spam.csv', encoding='ISO-8859-1')
```

```python
punctuationfree = ' '
def remove_punctuation(text):
  punctuationfree = ''.join([i for i in text if i not in string.punctuation])
  return punctuationfree
```

```python
remove_punctuation('he###llo!!')
```

```python
data['clean-msg']=data['v2'].apply(lambda x: remove_punctuation(x))
data['clean-msg']=data['clean-msg'].apply(lambda x: x.lower())
```

```python
import nltk
def tokenization(text):
  words = nltk.word_tokenize(text)
  return words
```

```python
nltk.download('punkt_tab')
```

```python
data['msg_tokenized'] = data['clean-msg'].apply(lambda x: tokenization(x))
```

```python
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
  output = ' '.join([i for i in text if i not in stopwords])
  return output
```

```python
data['no_stopwords'] = data['msg_tokenized'].apply(lambda x:remove_stopwords(x))
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

```

```python
def stemming(text):
  stem=[stemmer.stem(word) for word in text]
  return stem
```

```python
data['clean-msg']=data['clean-msg'].apply(lambda x: stemming(x))
```

```python
documents = ["This is the first document.",
             "This document is the second document."
             "And this is the third one."]

vocabulary = set()
for document in documents:
  for word in document.lower().split():
    vocabulary.add (word)
bow_representation = []
for document in documents:
  document_bow = []
  for word in vocabulary:
    if word in document.lower().split():
      document_bow.append(1)
    else:
      document_bow.append(0)
  bow_representation.append(document_bow)
print ("Vocabulary:", list(vocabulary))
for i, doc_bow in enumerate (bow_representation):
  print (f"Document {i+1}:", doc_bow)
```
