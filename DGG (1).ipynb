```python
import tensorflow as tf
from keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPool2D, Flatten, Dropout, ReLU, Softmax
```

```python
vgg_layers= [ Input((224, 224, 3)),
             Conv2D(filters=64, kernel_size=3, padding='same', strides=1, activation='relu', name='layer1'),
             Conv2D(filters=64, kernel_size=3, padding='same', strides=1, activation='relu', name='layer2'),
             MaxPool2D(pool_size=2, name='MaxPool1'),

             Conv2D(filters=128, kernel_size=3, padding='same', strides=1, activation='relu', name='layer3'),
             Conv2D(filters=128, kernel_size=3, padding='same', strides=1, activation='relu', name='layer4'),
             MaxPool2D(pool_size=2, name='MaxPool2'),

             Conv2D(filters=256, kernel_size=3, padding='same', strides=1, activation='relu', name='layer5'),
             Conv2D(filters=256, kernel_size=3, padding='same', strides=1, activation='relu', name='layer6'),
             Conv2D(filters=256, kernel_size=3, padding='same', strides=1, activation='relu', name='layer7'),
             MaxPool2D(pool_size=2, name='MaxPool3'),

             Conv2D(filters=512, kernel_size=3, padding='same', strides=1, activation='relu', name='layer8'),
             Conv2D(filters=512, kernel_size=3, padding='same', strides=1, activation='relu', name='layer9'),
             Conv2D(filters=512, kernel_size=3, padding='same', strides=1, activation='relu', name='layer10'),
             MaxPool2D(pool_size=2, name='MaxPool4'),

             Conv2D(filters=512, kernel_size=3, padding='same', strides=1, activation='relu', name='layer11'),
             Conv2D(filters=512, kernel_size=3, padding='same', strides=1, activation='relu', name='layer12'),
             Conv2D(filters=512, kernel_size=3, padding='same', strides=1, activation='relu', name='layer13'),
             MaxPool2D(pool_size=2, name='MaxPool5'),

             Flatten(),

             Dense(4096,activation='relu',name='layer14'),
             Dropout(0.5),

             Dense(4096,activation='relu',name='layer15'),
             Dropout(0.5),

             Dense(1000,activation='relu',name='layer16'),
             Softmax (name='softmax')

          ]
```

```python
model = Sequential(vgg_layers)
model.summary()
```

```python
x = tf.random.normal((1,224,224,3))
out = model(x)
print(out.shape)
```
